```python
try:
    # Create a slice
    slice = fablib.new_slice(name="NDNSlice")

    # Design the slice
    node_c1 = slice.add_node(name="c1", image='default_ubuntu_18', site='FIU', disk=30) #MAX DOWN
    node_c2 = slice.add_node(name="c2", image='default_ubuntu_18', site='FIU', disk=30) 
    node_c3 = slice.add_node(name="c3", image='default_ubuntu_18', site='FIU', disk=30) 
    node_c4 = slice.add_node(name="c4", image='default_ubuntu_18', site='FIU', disk=30) 
    node_c5 = slice.add_node(name="c5", image='default_ubuntu_18', site='FIU', disk=30) 

    node_p = slice.add_node(name="p",  image='default_ubuntu_18', site='FIU', disk=30)
    node_r1 = slice.add_node(name="r1",  image='default_ubuntu_18', site='FIU', disk=30)
   

    iface_c1 = node_c1.add_component(model='NIC_Basic', name='c1_nic').get_interfaces()[0]
    iface_c2 = node_c2.add_component(model='NIC_Basic', name='c2_nic').get_interfaces()[0]
    iface_c3 = node_c3.add_component(model='NIC_Basic', name='c3_nic').get_interfaces()[0]
    iface_c4 = node_c4.add_component(model='NIC_Basic', name='c4_nic').get_interfaces()[0]
    iface_c5 = node_c5.add_component(model='NIC_Basic', name='c5_nic').get_interfaces()[0]

    iface_p = node_p.add_component(model='NIC_Basic', name='p_nic').get_interfaces()[0]
    iface_r1c = node_r1.add_component(model='NIC_Basic', name='r1_c_nic').get_interfaces()[0]
    # iface_r1c2 = node_r1.add_component(model='NIC_Basic', name='r1_c2_nic').get_interfaces()[0]
    # iface_r1c3 = node_r1.add_component(model='NIC_Basic', name='r1_c3_nic').get_interfaces()[0]
    # iface_r1c4 = node_r1.add_component(model='NIC_Basic', name='r1_c4_nic').get_interfaces()[0]
    # iface_r1c5 = node_r1.add_component(model='NIC_Basic', name='r1_c5_nic').get_interfaces()[0]

    iface_r1p = node_r1.add_component(model='NIC_Basic', name='r1_p_nic').get_interfaces()[0]

    net1 = slice.add_l2network(name='r1cnet', interfaces=[iface_c1, iface_r1c, iface_c2, iface_c3, iface_c4, iface_c5], type='L2Bridge')
    # net1 = slice.add_l2network(name='r1c2net', interfaces=[iface_c2, iface_r1c2], type='L2Bridge')
    # net1 = slice.add_l2network(name='r1c3net', interfaces=[iface_c3, iface_r1c3], type='L2Bridge')
    # net1 = slice.add_l2network(name='r1c4net', interfaces=[iface_c4, iface_r1c4], type='L2Bridge')
    # net1 = slice.add_l2network(name='r1c5net', interfaces=[iface_c5, iface_r1c5], type='L2Bridge')

    net2 = slice.add_l2network(name='r1pnet', interfaces=[iface_p, iface_r1p], type='L2Bridge')

    #Submit the Request
    slice.submit()
except Exception as e:
    print(f"Exception: {e}")
```

```python
slice = fablib.get_slice("NDNSlice")

try:
    for node in slice.get_nodes():
        print(f"{node}")
        print(f"ssh -F ~/work/fabric_config/ssh_config {node.get_username()}@{node.get_management_ip()}")
except Exception as e:
    print(f"Exception: {e}")
```

```python
consumers = [
    {'name':'c1', 'net':'192.168.1.0', 'ip':'192.168.1.2', 'router_ip':'192.168.1.1'},
    {'name':'c2', 'net':'192.168.1.0', 'ip':'192.168.1.3', 'router_ip':'192.168.1.1'},
    {'name':'c3', 'net':'192.168.1.0', 'ip':'192.168.1.4', 'router_ip':'192.168.1.1'},
    {'name':'c4', 'net':'192.168.1.0', 'ip':'192.168.1.5', 'router_ip':'192.168.1.1'},
    {'name':'c5', 'net':'192.168.1.0', 'ip':'192.168.1.6', 'router_ip':'192.168.1.1'}]
routers = [
        {'name':'r1'}
]

producer = {'name':'p', 'net':'192.168.0.0', 'ip':'192.168.0.2', 'router_ip':'192.168.0.1'}

producer['node'] = slice.get_node(producer['name']);
producer['usr'] = str(producer['node'].get_username())
producer['man_ip'] = str(producer['node'].get_management_ip())
producer['if'] = producer['node'].get_interfaces()[0].get_os_interface()

for r in routers:
    r['node'] = slice.get_node(r['name'])
    r['ifs'] =r['node'].get_interfaces()
    r['man_ip'] = str(r['node'].get_management_ip())
    r['usr'] = str(r['node'].get_username())
    for interface in r['ifs']:
        if interface.get_name() == f"{r['name']}-{r['name']}_p_nic-p1":
            producer['router_if'] = interface
            producer['router_if_name'] = interface.get_os_interface()
            producer['router'] = r

    
for c in consumers:
    c['node'] = slice.get_node(c['name'])
    c['man_ip'] = str(c['node'].get_management_ip())
    c['if'] = str(c['node'].get_interfaces()[0].get_os_interface())
    c['usr'] = str(c['node'].get_username())
    c['router'] = routers[0]
    c['router_name']=c['router']['name']
    
    for interface in c['router']['ifs']:
        # print(interface.get_name())
        if interface.get_name() == f"{c['router_name']}-{c['router_name']}_{c['name']}_nic-p1" or interface.get_name() == f"{c['router_name']}-{c['router_name']}_c_nic-p1":
            c['router_if'] = interface 
            c['router_if_name'] = interface.get_os_interface()

print(f'{producer}')
print(f'{consumers}')
print(f'{routers}')

# print(client.execute("echo hello"))
# print(server.execute("echo hi"))
```

```python
producer['node'].execute(f"sudo ip addr add {producer['ip']}/24 dev {producer['if']}")
print(producer['node'].execute(f"ip addr show dev {producer['if']}"))
print(producer['node'].execute(f"ip route show"))

print()
producer['router']['node'].execute(f"sudo ip addr add {producer['router_ip']}/24 dev {producer['router_if_name']}")
for r in routers:
    router = r['node']
    router.execute("sudo sysctl -w net.ipv4.ip_forward=1")
    print(router.execute("ip route show"))
    print()
    
for c in consumers:
    node = c['node']
    p_node = producer['node']
    router = c['router']['node']

    node.execute(f"sudo ip addr add {c['ip']}/24 dev {c['if']}")
    router.execute(f"sudo ip addr add {c['router_ip']}/24 dev {c['router_if_name']}")
    node.execute(f"sudo ip route add {producer['net']}/24 via {c['router_ip']} dev {c['if']}")

    print(node.execute(f"ip addr show dev {c['if']}"))
    print(node.execute("ip route show"))
    print()
    
    p_node.execute(f"sudo ip route add {c['net']}/24 via {producer['router_ip']} dev {producer['router_if_name']}")
    print(router.execute(f"ip addr show dev {c['router_if_name']}"))
    print()
print()




```
